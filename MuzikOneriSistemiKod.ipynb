```python
# Gerekli Kütüphaneler
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# 1. VERİ
dosya_yolu = '/content/drive/MyDrive/VeriSeti/Genel_Veriseti.csv'
df = pd.read_csv(dosya_yolu)

# Temizlik
df = df[df['popularity'] > 5]
df['song_age'] = 2025 - df['release_year']
features = ['tempo', 'energy', 'danceability', 'loudness', 'valence',
            'acousticness', 'instrumentalness', 'speechiness', 'song_age', 'artist_popularity']

df_clean = df.dropna(subset=features + ['popularity'])
X = df_clean[features]
y = df_clean['popularity']

# Ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

#DEEP LEARNING
print(" Yapay Sinir Ağı ")

model_dl = Sequential()

# Giriş Katmanı + 1. Gizli Katman (64 Nöron)
model_dl.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
# Dropout: Ezberlemeyi önlemek için nöronların %20'sini rastgele kapatır
model_dl.add(Dropout(0.2))

# 2. Gizli Katman (32 Nöron)
model_dl.add(Dense(32, activation='relu'))

# 3. Gizli Katman (16 Nöron)
model_dl.add(Dense(16, activation='relu'))

# Çıkış Katmanı (Tek bir sayı tahmin edeceğimiz için 1 nöron)
model_dl.add(Dense(1, activation='linear'))

# Modeli Derle (Compile)
# Optimizer: Adam (En popüler öğrenme algoritması)
# Loss: Mean Absolute Error (Hata payını minimize etmeye çalış)
model_dl.compile(optimizer='adam', loss='mean_absolute_error')

#
print(" Model Eğitiliyor ")
history = model_dl.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    verbose=0
)

print(" Eğitim Tamamlandı!")

# --- 4. SONUÇLAR ---
y_pred = model_dl.predict(X_test)

# R2 ve MAE Hesapla
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("\n" + "="*30)
print(" DEEP LEARNING SONUÇLARI")
print("="*30)
print(f"R2 Skoru: {r2:.4f}")
print(f"Hata Payı (MAE): {mae:.2f}")
print("="*30)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error


cols = [
    'tempo', 'energy', 'danceability', 'loudness', 'valence',
    'acousticness', 'instrumentalness', 'speechiness',
    'sarki_yasi', 'artist_popularity'
]

# Eğer 'sarki_yasi' veya 'artist_popularity' df'de yoksa, hata almamak için üretelim:
if 'sarki_yasi' not in df.columns:
    df['sarki_yasi'] = 2025 - df['release_year']

if 'artist_popularity' not in df.columns:
    # Basit bir artist popularity hesaplaması
    artist_pop_map = df.groupby('artist')['popularity'].mean().to_dict()
    df['artist_popularity'] = df['artist'].map(artist_pop_map).fillna(50)

# ---------------------------------------------------------
# 2. ZAMANSAL BÖLME (TEMPORAL SPLIT)
# ---------------------------------------------------------
# İsteğine uygun olarak 2020 yılını test, öncesini eğitim yapalım.
test_yili = 2020
egitim_bitis_yili = 2019

print(f"\n ZAMANSAL TEST SENARYOSU")
print(f"   Eğitim Dönemi: 19XX - {egitim_bitis_yili}")
print(f"   Test Dönemi (Gelecek Simülasyonu): {test_yili}")

# Maskeleme (Veriyi yıllara göre ayırma)
train_mask = df['release_year'] <= egitim_bitis_yili
test_mask = df['release_year'] == test_yili

# Eğitim ve Test verilerini ayır
X_train_time = df.loc[train_mask, cols]
y_train_time = df.loc[train_mask, 'popularity']

X_test_time = df.loc[test_mask, cols]
y_test_time = df.loc[test_mask, 'popularity']

print(f"   Eğitim Seti Boyutu: {len(X_train_time)} şarkı")
print(f"   Test Seti Boyutu: {len(X_test_time)} şarkı")

# ---------------------------------------------------------
# 3. MODEL EĞİTİMİ
# ---------------------------------------------------------
print("\n⚙️  Model geçmiş verilerle eğitiliyor...")
model_time = XGBRegressor(
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=6,
    random_state=42,
    n_jobs=-1
)

model_time.fit(X_train_time, y_train_time)

# ---------------------------------------------------------
# 4. GELECEĞİ TAHMİN ETME
# ---------------------------------------------------------
preds_time = model_time.predict(X_test_time)

# ---------------------------------------------------------
# 5. SONUÇLARI RAPORLAMA
# ---------------------------------------------------------
r2_time = r2_score(y_test_time, preds_time)
mae_time = mean_absolute_error(y_test_time, preds_time)

print("\n" + "="*50)
print(f"🚀 GELECEK SİMÜLASYONU SONUCU ({test_yili} Yılı)")
print("="*50)
print(f"✅ R2 Başarısı: {r2_time:.4f}")
print(f"📉 Ortalama Hata Payı (MAE): {mae_time:.2f}")
print("="*50)

# ---------------------------------------------------------
# 6. KANIT GRAFİĞİ
# ---------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test_time, y=preds_time, alpha=0.6, color='purple', s=80, edgecolor='w')
plt.plot([0, 100], [0, 100], 'r--', lw=3, label='Mükemmel Tahmin Çizgisi')
plt.title(f"ZAMANSAL DOĞRULAMA: {test_yili} Yılı Tahminleri\n(Model bu yılı hiç görmeden tahmin etti)", fontsize=14, fontweight='bold')
plt.xlabel(f'Gerçek {test_yili} Popülaritesi')
plt.ylabel('Model Tahmini')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
import warnings

# 1. UYARILARI KAPAT
warnings.filterwarnings('ignore')

# 2. VERIYI YUKLE VE HAZIRLA
path = '//content/drive/MyDrive/VeriSeti/Genel_Veriseti.csv'
df = pd.read_csv(path)

# Veri Temizligi
df = df[df['popularity'] > 5]
df['sarki_yasi'] = 2025 - df['release_year']
cols = ['tempo', 'energy', 'danceability', 'loudness', 'valence',
        'acousticness', 'instrumentalness', 'speechiness', 'sarki_yasi', 'artist_popularity']

df_final = df.dropna(subset=cols + ['popularity'])
X = df_final[cols]
y = df_final['popularity']

# Standardizasyon
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 3. MODEL EGITIMI
print("Modeller hazirlaniyor, lutfen bekleyin...")

xgboost_model = XGBRegressor(n_estimators=1000, learning_rate=0.05, max_depth=6, random_state=42)
xgboost_model.fit(X_train, y_train)

print("Model egitimi tamamlandi.")

# ==========================================================
# 4. SADE ARAYUZ (SIMULATOR)
# ==========================================================
print("\n" + "="*50)
print("2025 HIT SARKI TAHMIN SISTEMI")
print("="*50)

style = {'description_width': '120px'}
layout_slider = widgets.Layout(width='400px')

# --- GRUP 1: GENEL BILGILER ---
dropdown_genre = widgets.Dropdown(
    options=['Seciniz...', 'Pop', 'Hip-Hop/Rap', 'Rock/Metal', 'Akustik/Indie', 'Elektronik/Dans'],
    value='Seciniz...',
    description='Sarki Turu:',
    style=style, layout=layout_slider
)
slider_artist = widgets.IntSlider(value=50, min=0, max=100, description='Sanatci Unu:', style=style, layout=layout_slider)
slider_year = widgets.IntSlider(value=2025, min=2020, max=2025, description='Cikis Yili:', style=style, layout=layout_slider)

# --- GRUP 2: SES OZELLIKLERI ---
slider_tempo = widgets.IntSlider(value=120, min=60, max=200, description='Tempo (BPM):', style=style, layout=layout_slider)
slider_energy = widgets.FloatSlider(value=0.5, min=0.0, max=1.0, step=0.01, description='Enerji:', style=style, layout=layout_slider)
slider_dance = widgets.FloatSlider(value=0.5, min=0.0, max=1.0, step=0.01, description='Dans:', style=style, layout=layout_slider)
slider_loud = widgets.FloatSlider(value=-10.0, min=-60.0, max=0.0, description='Ses (dB):', style=style, layout=layout_slider)

# --- GRUP 3: DETAYLAR ---
slider_acoustic = widgets.FloatSlider(value=0.1, min=0.0, max=1.0, step=0.01, description='Akustik:', style=style, layout=layout_slider)
slider_valence = widgets.FloatSlider(value=0.5, min=0.0, max=1.0, description='Mutluluk:', style=style, layout=layout_slider)
slider_speech = widgets.FloatSlider(value=0.05, min=0.0, max=1.0, description='Soz Orani:', style=style, layout=layout_slider)
slider_instru = widgets.FloatSlider(value=0.0, min=0.0, max=1.0, description='Enstrumantal:', style=style, layout=layout_slider)

# CIKTI ALANI
out = widgets.Output(layout={'border': '1px solid #ccc', 'padding': '10px', 'margin': '10px 0'})

# BUTON
btn_predict = widgets.Button(
    description="ANALIZ ET",
    button_style='primary', # Mavi sade buton
    layout=widgets.Layout(width='400px', height='40px')
)

# TUR AYARLARI GUNCELLEME
def tur_degisti(change):
    tur = change['new']
    if tur == 'Pop':
        slider_energy.value = 0.75; slider_dance.value = 0.80; slider_acoustic.value = 0.1; slider_loud.value = -5.0
    elif tur == 'Rock/Metal':
        slider_energy.value = 0.95; slider_dance.value = 0.40; slider_acoustic.value = 0.0; slider_loud.value = -3.0
    elif tur == 'Akustik/Indie':
        slider_energy.value = 0.30; slider_dance.value = 0.40; slider_acoustic.value = 0.90; slider_loud.value = -12.0
    elif tur == 'Hip-Hop/Rap':
        slider_energy.value = 0.65; slider_dance.value = 0.85; slider_speech.value = 0.40
    elif tur == 'Elektronik/Dans':
        slider_energy.value = 0.90; slider_dance.value = 0.90; slider_instru.value = 0.60; slider_tempo.value = 128

dropdown_genre.observe(tur_degisti, names='value')

# TAHMIN FONKSIYONU
def tahmin_et(b):
    with out:
        clear_output(wait=True)

        yeni_sarki = np.array([[
            slider_tempo.value, slider_energy.value, slider_dance.value, slider_loud.value,
            slider_valence.value, slider_acoustic.value, slider_instru.value, slider_speech.value,
            2025 - slider_year.value, slider_artist.value
        ]])

        yeni_sarki_scaled = scaler.transform(yeni_sarki)
        puan = xgboost_model.predict(yeni_sarki_scaled)[0]

        print(f"SECILEN TUR: {dropdown_genre.value}")
        print("-" * 30)
        print(f"TAHMINI BASARI SKORU: {puan:.1f} / 100")

        if puan > 80:
            print("\nSONUC: KESIN HIT (Global Listeler)")
        elif puan > 65:
            print("\nSONUC: BASARILI (Ulusal Listeler)")
        elif puan > 50:
            print("\nSONUC: ORTALAMA (Radyo Sarkisi)")
        elif puan > 30:
            print("\nSONUC: NIS KITLE (Alternatif)")
        else:
            print("\nSONUC: RISKLI (Ticari Basari Zor)")
        print("-" * 30)

btn_predict.on_click(tahmin_et)

# ARAYUZU GOSTER
ui = widgets.VBox([
    widgets.HTML("<b>1. Sanatci ve Tur Bilgileri</b>"),
    dropdown_genre, slider_artist, slider_year,

    widgets.HTML("<b>2. Ses Ozellikleri</b>"),
    slider_tempo, slider_energy, slider_dance, slider_loud,

    widgets.HTML("<b>3. Diger Detaylar</b>"),
    slider_acoustic, slider_valence, slider_speech, slider_instru,

    widgets.HTML("<br>"),
    btn_predict,
    out
])

display(ui)
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import scipy.sparse as sp
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
from google.colab import drive

# 1. DRIVE BAĞLANTISI
drive.mount('/content/drive')

# 2. VERİ YÜKLEME (Drive Yoluna Göre Ayarlandı)
# sep=None ve engine='python' kullanarak ayırıcıyı (virgül veya noktalı virgül) otomatik algılamasını sağladık.
path = '/content/drive/MyDrive/VeriSeti/Genel_Veriseti.csv'
df = pd.read_csv(path, sep=None, engine='python')

print("Toplam Satır:", len(df))

# Gerekli Olan Özellikler (Features)
features = [
    'danceability','energy','valence','acousticness','instrumentalness',
    'tempo','loudness','speechiness','liveness','popularity','release_year'
]

# Eşleştirme
cols_match = {}

for f in features:
    found = False
    for c in df.columns:
        clean_col = c.lower().strip()
        if clean_col == f.lower():
            cols_match[f] = c
            found = True
            break

    if not found:
        print(f"ERROR: '{f}' bulunamadı!")

print("\nVeri temizleniyor ve sayıya dönüştürülüyor...")

for feature_name, csv_col_name in cols_match.items():
    # Virgülü nokta yap, string'e çevir
    df[csv_col_name] = df[csv_col_name].astype(str).str.replace(',', '.', regex=False)

    # 'nan'-> NaN
    df[csv_col_name] = df[csv_col_name].replace('nan', np.nan)

    # Sayısal değere floata dönüştürülür
    df[csv_col_name] = pd.to_numeric(df[csv_col_name], errors='coerce')

    if csv_col_name != feature_name:
        df.rename(columns={csv_col_name: feature_name}, inplace=True)

# Verilerin Kontrolü
existing_features = [f for f in features if f in df.columns]

print("\n--- Sütun Tipleri (float) ---")
print(df[existing_features].dtypes)

print("\n--- İlk 5 Satır Features Örneği ---")
print(df[existing_features].head())

# Genre kolonu bulma ve temizleme
genre_col = None
for c in df.columns:
    if c.lower() == 'genre':
        genre_col = c
        break
if genre_col is None:
    # Hata fırlatmak yerine uyarı verip boş oluşturuyoruz, kodun akışı bozulmasın
    print("UYARI: 'genre' sütunu bulunamadı, 'unknown' olarak devam ediliyor.")
    df['genre'] = 'unknown'
    genre_col = 'genre'
else:
    df[genre_col] = df[genre_col].fillna('unknown').astype(str)

numeric_cols_present = [cols_match[k] for k in cols_match if k in df.columns]
print("Kullanılacak numeric kolonlar:", numeric_cols_present)

df_numeric = df[numeric_cols_present].copy()
# Basit eksik değer doldurma: kolon ortalaması
df_numeric = df_numeric.fillna(df_numeric.mean())

# Genre için TF-IDF (Tür Verisi) encoding
tfidf = TfidfVectorizer(token_pattern=r"(?u)\b\w[\w+\s&-]*\b", lowercase=True)
genre_tfidf = tfidf.fit_transform(df[genre_col])
print("Genre TF-IDF:", genre_tfidf.shape)

# Sayısal özellikleri ölçekleme
scaler = StandardScaler()
numeric_scaled = scaler.fit_transform(df_numeric)

# Birleştirilmiş vektör (sparse) oluşturma (Sayısal Özellikler + Müzik Türü) combine etme.
numeric_sparse = sp.csr_matrix(numeric_scaled)
combined = sp.hstack([numeric_sparse, genre_tfidf], format='csr')
print("Combined vector shape:", combined.shape)

# Mod metrikleri
mood_set = {
    'mutlu': {'valence': (0.6, 1.0), 'energy': (0.4, 1.0)},
    'üzgün': {'valence': (0.0, 0.4), 'energy': (0.0, 0.5)},
    'enerjik': {'valence': (0.3, 1.0), 'energy': (0.6, 1.0)},
    'rahat': {'valence': (0.3, 0.8), 'energy': (0.0, 0.5)},
    'nötr': {'valence': (0.4, 0.6), 'energy': (0.3, 0.6)},
    'tümü': {'valence': (0.0, 1.0), 'energy': (0.0, 1.0)},
}

# Genre kolonunu string yap
df["genre"] = df["genre"].fillna("").astype(str)

# Tüm türleri parçala
all_genres = (
    df["genre"]
    .str.lower()
    .str.split(",")
    .explode()
    .str.strip()
    .unique()
)

print("GENRE Sayısı:", len(all_genres))
# print(all_genres) # Uzun çıktı olmaması için kapalı, istersen açabilirsin

genre_counts = (
    df["genre"]
    .str.lower()
    .str.split(",")
    .explode()
    .str.strip()
    .value_counts()
)
print("Frekansa Göre Genre (En Sık Dinlenilenler):")
print(genre_counts.head(20))

# Öneri Fonksiyonu
def recommend_name(song_name=None, artist_name=None, top_n=10, genre_filter=None, mood='tümü'):

    # QUERY VEKTÖRÜ OLUŞTUR
    if song_name:
        if artist_name:
            match = df[
                (df["name"].astype(str).str.lower() == song_name.lower()) &
                (df["artist"].astype(str).str.lower() == artist_name.lower())
            ]
            if match.empty:
                match = df[
                    df["name"].astype(str).str.lower().str.contains(song_name.lower(), na=False) &
                    df["artist"].astype(str).str.lower().str.contains(artist_name.lower(), na=False)
                ]
        else:
            match = df[df["name"].astype(str).str.lower() == song_name.lower()]
            if match.empty:
                match = df[df["name"].astype(str).str.lower().str.contains(song_name.lower(), na=False)]

        if match.empty:
            print(f"UYARI: Şarkı '{song_name}' bulunamadı.")
            return pd.DataFrame()

        idx = match.index[0]
        query_vec = combined[idx]   # 1 x N sparse vector

    elif artist_name:
        artist_songs = df[df["artist"].astype(str).str.lower() == artist_name.lower()]

        if artist_songs.empty:
            print(f"UYARI: Artist '{artist_name}' bulunamadı.")
            return pd.DataFrame()

        # Sparse ortalama
        query_vec = combined[artist_songs.index].mean(axis=0)
        query_vec = sp.csr_matrix(query_vec)

    else:
        raise ValueError("En az song_name veya artist_name verilmelidir")


    # MOOD FİLTRESİ
    mood_key = mood if mood in mood_set else 'tümü'
    vmin, vmax = mood_set[mood_key]["valence"]
    emin, emax = mood_set[mood_key]["energy"]

    mood_mask = (
        (df["valence"] >= vmin) & (df["valence"] <= vmax) &
        (df["energy"] >= emin) & (df["energy"] <= emax)
    )
    adaylar = df[mood_mask].index

    # GENRE FİLTRESİ
    if genre_filter:
        genre_mask = df[genre_col].str.contains(genre_filter, case=False, na=False)
        adaylar = df[genre_mask & df.index.isin(adaylar)].index


    # COSINE SIMILARITY
    cand_matrix = combined[adaylar]

    # Eğer aday yoksa boş dön
    if cand_matrix.shape[0] == 0:
        return pd.DataFrame()

    sims = cosine_similarity(query_vec, cand_matrix).flatten()

    sorted_idx = np.argsort(sims)[::-1]
    adaylar = np.array(adaylar)


    # SONUÇLAR
    results = []
    for i in sorted_idx:
        song_i = adaylar[i]

        if song_name and 'idx' in locals() and song_i == idx:
            continue

        results.append({
            "name": df.loc[song_i, "name"],
            "artist": df.loc[song_i, "artist"],
            "genre": df.loc[song_i, genre_col],
            "valence": df.loc[song_i, "valence"],
            "energy": df.loc[song_i, "energy"],
            "similarity": round(float(sims[i]), 4)
        })

        if len(results) >= top_n:
            break

    return pd.DataFrame(results)

# Testler (Cosine Similarity)
print("\n------ İlk Öneri Cosine Similarity (Sadece artist_name = Dua Lipa) ------")
recommend_1 = recommend_name(artist_name = "Dua Lipa")
print(recommend_1)

print("\n------ İkinci Öneri Cosine Similarity (artist_name = Rammstein | song_name = Sonne)-----")
recommend_2 = recommend_name(artist_name = "Rammstein", song_name = "Sonne")
print(recommend_2)

print("\n----- Ücüncü Öneri (artis_name = Rihanna | mood = enerjik | genre = pop) -------")
recommend_3 = recommend_name(artist_name = "Rihanna", genre_filter = 'pop' ,mood="enerjik")
print(recommend_3)

# PCA Bölümü
numeric_cols_pca = [
    "acousticness", "danceability", "energy", "instrumentalness",
    "liveness", "loudness", "speechiness", "tempo",
    "valence", "popularity", "release_year"
]

numeric_cols_pca = [c for c in numeric_cols_pca if c in df.columns]
print("\nKullanılan PCA numeric kolonları:", numeric_cols_pca)

# NA doldurma ve ölçekleme
df_numeric_pca = df[numeric_cols_pca].fillna(df[numeric_cols_pca].mean())

scaler_pca = StandardScaler()
numeric_scaled_pca = scaler_pca.fit_transform(df_numeric_pca)

print("Numeric scaled shape:", numeric_scaled_pca.shape)

# PCA (%95 varyans koruma)
pca = PCA(n_components=0.95)
pca_components = pca.fit_transform(numeric_scaled_pca)

print("PCA Output shape:", pca_components.shape)

def recommend_pca(song_name=None, artist_name=None, top_n=10, genre_filter=None, mood='tümü'):

    # QUERY VEKTÖRÜ OLUŞTUR
    if song_name:
        if artist_name:
            match = df[
                (df["name"].astype(str).str.lower() == song_name.lower()) &
                (df["artist"].astype(str).str.lower() == artist_name.lower())
            ]
            if match.empty:
                match = df[
                    df["name"].astype(str).str.lower().str.contains(song_name.lower(), na=False) &
                    df["artist"].astype(str).str.lower().str.contains(artist_name.lower(), na=False)
                ]
        else:
            match = df[df["name"].astype(str).str.lower() == song_name.lower()]
            if match.empty:
                match = df[df["name"].astype(str).str.lower().str.contains(song_name.lower(), na=False)]

        if match.empty:
            print(f"UYARI: Şarkı '{song_name}' bulunamadı.")
            return pd.DataFrame()

        idx = match.index[0]
        query_vec = pca_components[idx].reshape(1, -1)

    elif artist_name:
        artist_songs = df[df["artist"].astype(str).str.lower() == artist_name.lower()]

        if artist_songs.empty:
            print(f"UYARI: Artist '{artist_name}' bulunamadı.")
            return pd.DataFrame()

        query_vec = pca_components[artist_songs.index].mean(axis=0)
        query_vec = query_vec.reshape(1, -1)

    else:
        raise ValueError("En az song_name veya artist_name verilmelidir")


    # MOOD filtresi
    mood_key = mood if mood in mood_set else 'tümü'
    vmin, vmax = mood_set[mood_key]['valence']
    emin, emax = mood_set[mood_key]['energy']

    mood_mask = (
        (df["valence"] >= vmin) & (df["valence"] <= vmax) &
        (df["energy"] >= emin) & (df["energy"] <= emax)
    )

    candidate_idx = df[mood_mask].index

    # GENRE filtresi
    if genre_filter:
        gmask = df["genre"].str.contains(genre_filter, case=False, na=False)
        candidate_idx = df[gmask & df.index.isin(candidate_idx)].index

    # PCA similarity
    candidate_matrix = pca_components[candidate_idx]

    if candidate_matrix.shape[0] == 0:
        return pd.DataFrame()

    sims = cosine_similarity(query_vec, candidate_matrix).flatten()
    sorted_idx = np.argsort(sims)[::-1]

    candidate_idx = np.array(candidate_idx)
    results = []

    for i in sorted_idx:
        song_i = candidate_idx[i]

        if song_name and 'idx' in locals() and song_i == idx:
            continue

        results.append({
            "name": df.loc[song_i, "name"],
            "artist": df.loc[song_i, "artist"],
            "genre": df.loc[song_i, "genre"],
            "similarity": round(float(sims[i]), 4)
        })

        if len(results) >= top_n:
            break

    return pd.DataFrame(results)

# Testler (PCA)
print("\n----- Birinci Öneri PCA (artist_name = Metallica) -------")
pca_1 = recommend_pca(artist_name = "Metallica")
print(pca_1)

print("\n----- İkinci Öneri PCA (Sadece artist_name = Dua Lipa) -------")
pca_2 = recommend_pca(artist_name = "Dua Lipa")
print(pca_2)

print("\n----- Ücüncü Öneri PCA (artist_name = Pink Floyd, genre_filter = progressive rock, mood = üzgün) ------")
pca_3 = recommend_pca(artist_name = "Pink Floyd", mood = "üzgün", genre_filter = "progressive rock")
print(pca_3)
```

```python
import pandas as pd
import numpy as np
import scipy.sparse as sp
import ipywidgets as widgets
from google.colab import drive
from IPython.display import display, clear_output
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
import warnings


warnings.filterwarnings('ignore')

# 1. DRIVE BAGLANTISI VE VERI YUKLEME
print("Google Drive baglaniyor ve veri isleniyor...")
drive.mount('/content/drive')

path = '/content/drive/MyDrive/VeriSeti/Genel_Veriseti.csv'
df = pd.read_csv(path, sep=None, engine='python') # Ayiriciyi otomatik algila


# MODUL 1: ONERI SISTEMI HAZIRLIK


# Kolon temizligi ve eslesme
features_recsys = [
    'danceability','energy','valence','acousticness','instrumentalness',
    'tempo','loudness','speechiness','liveness','popularity','release_year'
]
cols_match = {}
for f in features_recsys:
    for c in df.columns:
        if c.lower().strip() == f.lower():
            cols_match[f] = c
            break

# Veri donusumleri
df_rec = df.copy()
for feature_name, csv_col_name in cols_match.items():
    df_rec[csv_col_name] = df_rec[csv_col_name].astype(str).str.replace(',', '.', regex=False)
    df_rec[csv_col_name] = df_rec[csv_col_name].replace('nan', np.nan)
    df_rec[csv_col_name] = pd.to_numeric(df_rec[csv_col_name], errors='coerce')
    if csv_col_name != feature_name:
        df_rec.rename(columns={csv_col_name: feature_name}, inplace=True)

# Genre isleme
genre_col = next((c for c in df_rec.columns if c.lower() == 'genre'), None)
if genre_col:
    df_rec[genre_col] = df_rec[genre_col].fillna('unknown').astype(str)
    tfidf = TfidfVectorizer(token_pattern=r"(?u)\b\w[\w+\s&-]*\b", lowercase=True)
    genre_tfidf = tfidf.fit_transform(df_rec[genre_col])
else:
    # Genre yoksa bos matris
    genre_tfidf = sp.csr_matrix((len(df_rec), 0))

# Sayisal veriler ve Olcekleme
numeric_cols_present = [cols_match[k] if cols_match[k] in df_rec.columns else k for k in cols_match]
df_numeric = df_rec[features_recsys].fillna(df_rec[features_recsys].mean())
scaler_rec = StandardScaler()
numeric_scaled = scaler_rec.fit_transform(df_numeric)

# Matris Birlestirme (Hybrid)
numeric_sparse = sp.csr_matrix(numeric_scaled)
combined_features = sp.hstack([numeric_sparse, genre_tfidf], format='csr')

# PCA Hesaplama
pca = PCA(n_components=0.95)
pca_components = pca.fit_transform(numeric_scaled)

# Mood Tanimlari
mood_set = {
    'Mutlu': {'valence': (0.6, 1.0), 'energy': (0.4, 1.0)},
    'Üzgün': {'valence': (0.0, 0.4), 'energy': (0.0, 0.5)},
    'Enerjik': {'valence': (0.3, 1.0), 'energy': (0.6, 1.0)},
    'Rahat': {'valence': (0.3, 0.8), 'energy': (0.0, 0.5)},
    'Nötr': {'valence': (0.4, 0.6), 'energy': (0.3, 0.6)},
    'Tümü': {'valence': (0.0, 1.0), 'energy': (0.0, 1.0)},
}


# MODUL 2: POPÜLERLİK SİSTEMİ


df_hit = df_rec.copy() # Temizlenmis veriyi kullan
df_hit = df_hit[df_hit['popularity'] > 5]
df_hit['sarki_yasi'] = 2025 - df_hit['release_year']
# artist_popularity veri setinde yoksa olustur
if 'artist_popularity' not in df_hit.columns:
    artist_pop = df_hit.groupby('artist')['popularity'].transform('mean')
    df_hit['artist_popularity'] = artist_pop

cols_hit = ['tempo', 'energy', 'danceability', 'loudness', 'valence',
            'acousticness', 'instrumentalness', 'speechiness', 'sarki_yasi', 'popularity']

# Artist popularity varsa ekle
if 'artist_popularity' in df_hit.columns:
    cols_train = cols_hit[:-1] + ['artist_popularity']
else:
    cols_train = cols_hit[:-1]

df_final_hit = df_hit.dropna(subset=cols_train + ['popularity'])
X = df_final_hit[cols_train]
y = df_final_hit['popularity']

scaler_hit = StandardScaler()
X_scaled = scaler_hit.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

xgboost_model = XGBRegressor(n_estimators=1000, learning_rate=0.05, max_depth=6, random_state=42)
xgboost_model.fit(X_train, y_train)

print("Sistem hazir. Arayuz yukleniyor...")


# ARAYUZ TASARIMI

# --- TAB 1: MUZIK ONERI SISTEMI ---
style_ui = {'description_width': '100px'}
layout_ui = widgets.Layout(width='95%')

txt_artist = widgets.Text(description='Sanatçı:', placeholder='Örn: Metallica', style=style_ui, layout=layout_ui)
txt_song = widgets.Text(description='Şarkı:', placeholder='Örn: Enter Sandman (Opsiyonel)', style=style_ui, layout=layout_ui)
drop_mood = widgets.Dropdown(options=list(mood_set.keys()), value='Tümü', description='Mood:', style=style_ui, layout=layout_ui)
drop_algo = widgets.Dropdown(options=['Karma (Genel)', 'PCA (Matematiksel)'], value='Karma (Genel)', description='Algoritma:', style=style_ui, layout=layout_ui)
btn_rec = widgets.Button(description='ÖNERİ GETİR', button_style='info', layout=widgets.Layout(width='100%'))
out_rec = widgets.Output(layout={'border': '1px solid #ddd', 'padding': '10px'})

def on_click_recommend(b):
    with out_rec:
        clear_output()
        artist = txt_artist.value.strip()
        song = txt_song.value.strip()
        mood = drop_mood.value
        use_pca = (drop_algo.value == 'PCA (Matematiksel)')

        if not artist and not song:
            print("Lütfen en az bir sanatçı veya şarkı ismi girin.")
            return

        try:
            # Filtreleme
            mood_data = mood_set[mood]
            vmin, vmax = mood_data['valence']
            emin, emax = mood_data['energy']

            mask = (df_rec["valence"] >= vmin) & (df_rec["valence"] <= vmax) & \
                   (df_rec["energy"] >= emin) & (df_rec["energy"] <= emax)
            candidates_idx = df_rec[mask].index

            # Sorgu Vektoru
            query_vec = None
            target_idx = -1

            # Isim/Artist eslestirme
            match = pd.DataFrame()
            if song:
                if artist:
                    match = df_rec[(df_rec["name"].astype(str).str.lower() == song.lower()) &
                                   (df_rec["artist"].astype(str).str.lower() == artist.lower())]
                if match.empty:
                    match = df_rec[df_rec["name"].astype(str).str.lower().str.contains(song.lower(), na=False)]
            elif artist:
                match = df_rec[df_rec["artist"].astype(str).str.lower() == artist.lower()]

            if match.empty:
                print("Eşleşen şarkı veya sanatçı bulunamadı.")
                return

            if use_pca:
                # PCA Algoritmasi
                if song and not match.empty:
                    target_idx = match.index[0]
                    query_vec = pca_components[target_idx].reshape(1, -1)
                else: # Sadece artist ortalamasi
                    query_vec = pca_components[match.index].mean(axis=0).reshape(1, -1)

                cand_matrix = pca_components[candidates_idx]
            else:
                # Hybrid Algoritma
                if song and not match.empty:
                    target_idx = match.index[0]
                    query_vec = combined_features[target_idx]
                else:
                    query_vec = combined_features[match.index].mean(axis=0)
                    query_vec = sp.csr_matrix(query_vec)

                cand_matrix = combined_features[candidates_idx]

            # Benzerlik
            sims = cosine_similarity(query_vec, cand_matrix).flatten()
            sorted_args = np.argsort(sims)[::-1]

            print(f"Sonuçlar ({drop_algo.value} - Mod: {mood}):\n")
            print(f"{'Şarkı':<40} {'Sanatçı':<30} {'Benzerlik'}")
            print("-" * 85)

            count = 0
            for i in sorted_args:
                idx = candidates_idx[i]
                if idx == target_idx: continue # Kendisini onerme

                s_row = df_rec.loc[idx]
                score = sims[i]
                print(f"{str(s_row['name'])[:38]:<40} {str(s_row['artist'])[:28]:<30} %{score*100:.1f}")

                count += 1
                if count >= 10: break

        except Exception as e:
            print(f"Bir hata oluştu: {str(e)}")

btn_rec.on_click(on_click_recommend)

tab1 = widgets.VBox([
    widgets.HTML("<b>Müzik Öneri Motoru</b>"),
    txt_artist, txt_song, drop_mood, drop_algo,
    widgets.HTML("<br>"), btn_rec, out_rec
])

#  TAB 2: POPÜLERLİK TAHMIN SISTEMI

# Widget Tanimlari
style_hit = {'description_width': '120px'}
layout_slider = widgets.Layout(width='400px')

dd_genre = widgets.Dropdown(options=['Seciniz...', 'Pop', 'Hip-Hop/Rap', 'Rock/Metal', 'Akustik/Indie', 'Elektronik/Dans'], value='Seciniz...', description='Sarki Turu:', style=style_hit, layout=layout_slider)
sl_artist = widgets.IntSlider(value=50, min=0, max=100, description='Sanatci Unu:', style=style_hit, layout=layout_slider)
sl_year = widgets.IntSlider(value=2025, min=2020, max=2025, description='Cikis Yili:', style=style_hit, layout=layout_slider)

sl_tempo = widgets.IntSlider(value=120, min=60, max=200, description='Tempo (BPM):', style=style_hit, layout=layout_slider)
sl_energy = widgets.FloatSlider(value=0.5, min=0.0, max=1.0, step=0.01, description='Enerji:', style=style_hit, layout=layout_slider)
sl_dance = widgets.FloatSlider(value=0.5, min=0.0, max=1.0, step=0.01, description='Dans:', style=style_hit, layout=layout_slider)
sl_loud = widgets.FloatSlider(value=-10.0, min=-60.0, max=0.0, description='Ses (dB):', style=style_hit, layout=layout_slider)

sl_acoustic = widgets.FloatSlider(value=0.1, min=0.0, max=1.0, step=0.01, description='Akustik:', style=style_hit, layout=layout_slider)
sl_valence = widgets.FloatSlider(value=0.5, min=0.0, max=1.0, description='Mutluluk:', style=style_hit, layout=layout_slider)
sl_speech = widgets.FloatSlider(value=0.05, min=0.0, max=1.0, description='Soz Orani:', style=style_hit, layout=layout_slider)
sl_instru = widgets.FloatSlider(value=0.0, min=0.0, max=1.0, description='Enstrumantal:', style=style_hit, layout=layout_slider)

out_hit = widgets.Output(layout={'border': '1px solid #ccc', 'padding': '10px'})
btn_predict = widgets.Button(description="ANALIZ ET", button_style='primary', layout=widgets.Layout(width='400px'))

# Tur Presetleri
def tur_update(change):
    t = change['new']
    if t == 'Pop':
        sl_energy.value=0.75; sl_dance.value=0.80; sl_acoustic.value=0.1; sl_loud.value=-5.0
    elif t == 'Rock/Metal':
        sl_energy.value=0.95; sl_dance.value=0.40; sl_acoustic.value=0.0; sl_loud.value=-3.0
    elif t == 'Akustik/Indie':
        sl_energy.value=0.30; sl_dance.value=0.40; sl_acoustic.value=0.90; sl_loud.value=-12.0
    elif t == 'Hip-Hop/Rap':
        sl_energy.value=0.65; sl_dance.value=0.85; sl_speech.value=0.40
    elif t == 'Elektronik/Dans':
        sl_energy.value=0.90; sl_dance.value=0.90; sl_instru.value=0.60; sl_tempo.value=128

dd_genre.observe(tur_update, names='value')

def predict_hit(b):
    with out_hit:
        clear_output(wait=True)
        # Modelin bekledigi sirada ozellik vektoru
        feats = [
            sl_tempo.value, sl_energy.value, sl_dance.value, sl_loud.value,
            sl_valence.value, sl_acoustic.value, sl_instru.value, sl_speech.value,
            2025 - sl_year.value
        ]
        #
        if 'artist_popularity' in cols_train:
            feats.append(sl_artist.value)

        vec = np.array([feats])
        vec_scaled = scaler_hit.transform(vec)
        score = xgboost_model.predict(vec_scaled)[0]

        print(f"TÜR: {dd_genre.value} | PUAN: {score:.1f}/100")
        print("-" * 40)
        if score > 80: print("SONUC: KESIN HIT (Global Listeler)")
        elif score > 65: print("SONUC: BASARILI (Ulusal Listeler)")
        elif score > 50: print("SONUC: ORTALAMA (Radyo Sarkisi)")
        elif score > 30: print("SONUC: NIS KITLE (Alternatif)")
        else: print("SONUC: RISKLI")

btn_predict.on_click(predict_hit)

tab2 = widgets.VBox([
    widgets.HTML("<b>1. Sanatci ve Tur</b>"), dd_genre, sl_artist, sl_year,
    widgets.HTML("<b>2. Ses Ozellikleri</b>"), sl_tempo, sl_energy, sl_dance, sl_loud,
    widgets.HTML("<b>3. Detaylar</b>"), sl_acoustic, sl_valence, sl_speech, sl_instru,
    widgets.HTML("<br>"), btn_predict, out_hit
])

#  ANA PENCERE
main_tabs = widgets.Tab(children=[tab1, tab2])
main_tabs.set_title(0, "Müzik Öneri")
main_tabs.set_title(1, "Hit Tahmincisi")

print("\n" + "="*40)
print("MÜZİK ANALİZ PLATFORMU")
print("="*40)
display(main_tabs)
```
